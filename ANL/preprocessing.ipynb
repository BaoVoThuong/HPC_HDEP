```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ANL-Intrepid-2009-1.swf.csv')
```

```python
df.describe()
```

```python
df = df.drop(['job_id','avg_cpu_time_used','used_memory','requested_memory','status','group_id','executable_id','partition_id','preceding_job_id','think_time'],axis =1 )
```

```python
df
```

```python
df.describe()
```

```python


corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

```
