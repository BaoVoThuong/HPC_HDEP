```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HCMUT-SuperNodeXP-2017-1.0.csv')
```

```python
df.describe()
```

```python
df = df.drop(['preceding_job_id','think_time','job_id','status'],axis= 1)
```

```python
corr_matrix = df.corr()

# Vẽ heatmap
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
